# DuckDB 101 - Local Intro

This notebook shows DuckDB running locally in VS Code.

The goal is to query local data files without setting up a server.

**Dataset:** US Airline On-Time Performance (BTS) — 2023 flights

```python
import duckdb
import pandas as pd
from pathlib import Path

DATA = Path("../data/flights")
```

```python
duckdb.sql("SELECT 'Hello from DuckDB' AS message").show()
```

```python
con = duckdb.connect()
```

```python
csv_path = DATA / "flights_2023.csv"

duckdb.sql(f"""
    SELECT *
    FROM read_csv_auto('{csv_path}')
    LIMIT 10
""").df()
```

```python
duckdb.sql(f"""
    SELECT
        Reporting_Airline,
        COUNT(*)                                       AS total_flights,
        ROUND(AVG(ArrDelay), 1)                        AS avg_arr_delay_min,
        ROUND(SUM(Cancelled) * 100.0 / COUNT(*), 1)   AS cancel_pct
    FROM read_csv_auto('{csv_path}')
    WHERE Cancelled = 0
    GROUP BY Reporting_Airline
    ORDER BY total_flights DESC
    LIMIT 10
""").df()
```

# Local Data Check

This notebook shows VS Code as a local data workbench.

Demo flow:

1. Read a local CSV file
2. Preview the data
3. Summarize the data
4. Create a quick chart

```python
import pandas as pd
from pathlib import Path
```

```python
csv_path = Path("../data/fallout_vaults.csv")

df = pd.read_csv(csv_path)
df.head()
```

```python
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]:,}")

df.columns
```

```python
df["Experiment_Category"].value_counts()
```

```python
df["Current_Status"].value_counts()
```

```python
df["Experiment_Category"].value_counts().plot(
    kind="bar",
    title="Vaults by Experiment Category"
)
```

## For a given workspace - get a list of semantic models 
## and iterate through each to check each report
## **<u>Can be expanded out to get list of workspaces<u>**
## **<u>and iterate through workspaces too</u>**

```python
import warnings
warnings.filterwarnings('ignore', category=FutureWarning, module='sempy_labs')
import sempy.fabric as fabric
import sempy_labs as labs
import pandas as pd

__workspace = 'Summit2025_SLL'  # The name or ID of the workspace in which the report exists

# Get list of models
models = fabric.list_datasets(__workspace)['Dataset Name'].to_list()

# Initialize list to store all results
all_invalid_objects = []
error_models = []

# Iterate through each model
for model in models:
    try:
        print(f"Processing: {model}...")
        df = labs.list_report_semantic_model_objects(dataset=model, workspace=__workspace, extended=True)
        
        # Filter for invalid objects
        invalid_objects = df[df['Valid Semantic Model Object'] == False].copy()
        
        if not invalid_objects.empty:
            # Add model name column for tracking
            invalid_objects.insert(0, 'Model Name', model)
            all_invalid_objects.append(invalid_objects)
            print(f"  ‚úó Found {len(invalid_objects)} invalid object(s)")
        else:
            print(f"  ‚úì All objects valid")
        
    except Exception as e:
        error_msg = f"{model}: {str(e)}"
        error_models.append(error_msg)
        print(f"  ‚úó Error: {str(e)}")
        continue

print("\n" + "="*80)
print("SUMMARY")
print("="*80)

# Display compiled results
if all_invalid_objects:
    print(f"\nüìã Invalid Objects Found ({sum(len(df) for df in all_invalid_objects)} total):\n")
    combined_df = pd.concat(all_invalid_objects, ignore_index=True)
    print(combined_df.to_string(index=False))
else:
    print("\n‚úì No invalid objects found in any model!")

# Display errors if any
if error_models:
    print(f"\n\n‚ö†Ô∏è  Errors Encountered ({len(error_models)}):\n")
    for error in error_models:
        print(f"  - {error}")

print("\n" + "="*80)
print(f"Total Models Processed: {len(models)}")
print(f"Models with Invalid Objects: {len(all_invalid_objects)}")
print(f"Models with Errors: {len(error_models)}")
print("="*80)
```
